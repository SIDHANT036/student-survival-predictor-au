```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.chdir('/Users/sidhantnarang/student-survival-predictor-au/student-survival-predictor-au')

df = pd.read_csv('data/synthetic/students.csv')

# 1. Stress rate by city
stress_by_city = df.groupby('city')['financial_stress'].mean()
stress_by_city.plot(kind='bar', title='Financial stress rate by city')
plt.ylabel('% stressed')
plt.tight_layout()
plt.savefig('reports/stress_by_city.png', dpi=150)
plt.show()

# 2. Correlation heatmap
fig, ax = plt.subplots(figsize=(9, 7))
sns.heatmap(df.select_dtypes('number').corr(), annot=True, fmt='.2f',
            cmap='coolwarm', ax=ax)
plt.title('Feature correlation matrix')
plt.tight_layout()
plt.savefig('reports/correlation_heatmap.png', dpi=150)
plt.show()

# 3. Savings distribution by stress label
df.groupby('financial_stress')['savings_weekly'].plot(
    kind='hist', alpha=0.6, bins=40, legend=True,
    title='Weekly savings: stressed vs stable'
)
plt.xlabel('Weekly savings ($)')
plt.tight_layout()
plt.savefig('reports/savings_distribution.png', dpi=150)
plt.show()
```
