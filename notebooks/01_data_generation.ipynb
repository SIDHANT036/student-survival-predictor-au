```python
import pandas as pd
import numpy as np
import os

# Set working directory to project root
os.chdir('/Users/sidhantnarang/student-survival-predictor-au/student-survival-predictor-au')
os.makedirs('data/synthetic', exist_ok=True)

np.random.seed(42)
n = 2000
cities = ['Melbourne', 'Sydney', 'Brisbane']

df = pd.DataFrame({
    'city': np.random.choice(cities, n),
    'rent_weekly': np.random.randint(200, 600, n),
    'income_weekly': np.random.randint(150, 700, n),
    'transport_weekly': np.random.randint(20, 60, n),
    'grocery_weekly': np.random.randint(50, 150, n),
    'tuition_weekly': np.random.randint(100, 350, n),
    'study_hours': np.random.randint(20, 50, n),
    'work_hours': np.random.randint(0, 20, n),
    'commute_hours': np.random.uniform(0.5, 3, n),
})

df['total_expenses'] = (df['rent_weekly'] + df['transport_weekly']
                        + df['grocery_weekly'] + df['tuition_weekly'])
df['savings_weekly'] = df['income_weekly'] - df['total_expenses']
df['financial_stress'] = (df['savings_weekly'] < 50).astype(int)

df.to_csv('data/synthetic/students.csv', index=False)
print(f"✅ Saved {len(df)} rows to data/synthetic/students.csv")
print(df.head())
print(f"\nStress rate: {df['financial_stress'].mean():.1%}")
```
